# label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
train_column=df2[['sex','smoker','region']]
train_column.head()
```

```python
rr=train_column.apply(le.fit_transform)
```

```python
rr.head()
```

```python
df2[['sex','smoker','region']] = rr[['sex','smoker','region']]
```

```python
df2.head()
```

# loop label encoding

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
df3.columns
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        df3[col]=le.fit_transform(df3[col])
```

```python
df3.head()
```

# one-hot

# one hot loop

```python
for col in df5.columns:
    if is_numeric_dtype(df5[col]):
        continue
    else:
        one_hot = pd.get_dummies(df5[col], drop_first=True)
        df5.drop(df5[[col]], axis=1,inplace=True)
        df5=pd.concat([df5,one_hot], axis=1)
```

```python
df5.head()
```

# ordinal

```python
from sklearn.preprocessing import OrdinalEncoder
```
